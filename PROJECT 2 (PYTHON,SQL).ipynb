# EXTRACTION FROM ZIP FILE

```python
import zipfile
zip_ref = zipfile.ZipFile("orders.csv.zip")
zip_ref.extractall()
zip_ref.close()
```

```python
import pandas as pd
df = pd.read_csv("orders.csv" ,  na_values=["Not Available","unknown"])
df["Ship Mode"].unique()
```

```python
df.columns=df.columns.str.upper()
df.columns=df.columns.str.replace(" ","_")
df.head(20)
```

```python
df.head(20)
df['DISCOUNT_PRICE']=df["LIST_PRICE"]*df["DISCOUNT_PERCENT"]*0.01
df["SELLING_PRICE"]=df["LIST_PRICE"]-df["DISCOUNT_PRICE"]
df["PROFIT"] =df["SELLING_PRICE"]-df["COST_PRICE"]
df
```

```python
df.dtypes
```

```python
##CONVERT OBJEECT DATA TYPE TO DATE DATA TYPE
df["ORDER_DATE"] = pd.to_datetime(df["ORDER_DATE"], format='%Y-%m-%d')
df
```

```python
##DELETING COLUMNS
df.drop(columns=["LIST_PRICE","COST_PRICE","DISCOUNT_PERCENT"],inplace=True)
```
